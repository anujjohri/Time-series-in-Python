# Naive approach

Naive Forecasting - Estimating technique in which the last period's actuals are used as this period's forecast, without adjusting them or attempting to establish causal factors. It is used only for comparison with the forecasts generated by the better (sophisticated) techniques.

```python
dd = np.asarray(train.Count)
y_hat = test.copy()
#print(dd)
y_hat['naive'] = dd[-1] #3408
#print(y_hat.head())
plt.figure(figsize=(10,6))
plt.plot(train.index , train['Count'] , label = 'Train')
plt.plot(test.index , test['Count'] , label = 'Test')
plt.plot(y_hat.index , y_hat['naive'] , label = 'Naive Forecast')
plt.legend(loc = 'best')
plt.title('Naive Forecast')
plt.show()
```

Calculating accuracy 

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_hat.Count , y_hat.naive))
print(rms)
```

# Moving average technique

Moving averages are a simple and common type of smoothing used in time series analysis and time series forecasting. Calculating a moving average involves creating a new series where the values are comprised of the average of raw observations in the original time series

```python
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['Count'].rolling(60).mean().iloc[-1] 



plt.figure(figsize=(10,6))
plt.plot(train.index , train['Count'] , label = 'Train')
plt.plot(test.index , test['Count'] , label = 'Test')
plt.plot( y_hat_avg['moving_avg_forecast'] , label = 'Moving Avg Forecast')
plt.legend(loc = 'best')
plt.title('Moving Avg Forecast')
plt.show()
```

Calculating accuracy 

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_hat_avg.Count , y_hat_avg.moving_avg_forecast))
print(rms)
```

# Exponential techiques

Exponential smoothing assigns exponentially decreasing weights as the observation get older. This is in contrast to single moving averages where the past observations are weighted equally. Exponential smoothing is a very popular scheme to produce a smoothed time series.

In the case of moving averages, the weights assigned to the observations are the same and are equal to 1/N. In exponential smoothing, however, there are one or more smoothing parameters to be determined (or estimated) and these choices determine the weights assigned to the observations.

# Simple exponential smoothing

Single Exponential Smoothing, SES for short, also called Simple Exponential Smoothing, is a time series forecasting method for univariate data without a trend or seasonality. It requires a single parameter, called alpha (a), also called the smoothing factor or smoothing coefficient.

range of alpha -> 0 to 1

low value - older value are equally important

high value - older value are less important 


simple exponential smoothing

consider entire data , focus more on recent datapoints -> assign weights in background

more suited where we dont have any trend or pattern - simple exp

if we have trend - better method halt linear (double exp)

if we have trend and seasonality - halt winter (triple exp)


```python
import statsmodels.api as sm
sm.tsa.seasonal_decompose(train.Count).plot()
plt.show()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
```

```python

y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['Count'],dtype = np.float)).fit(smoothing_level = 0.1) 

y_hat_avg['SES'] = fit2.forecast(len(test))

plt.figure(figsize=(10,6))
plt.plot( train['Count'] , label = 'Train')
plt.plot(test['Count'] , label = 'Test')
plt.plot( y_hat_avg['SES'] , label = 'SES')
plt.legend(loc = 'best')
plt.title('SES')
plt.show()
```

Calculation accuracy

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_hat_avg.Count , y_hat_avg.SES))
print(rms)
```

# Double exponential smoothing

Double Exponential Smoothing is an improvement of Simple Exponential Smoothing, also known as Exponential Moving Average, which does the exponential filter process twice. It's usually been used to predict the future data in time series analysis, where there is a trend in the data.

Double exponential smoothing employs a level component and a trend component at each period. Double exponential smoothing uses two weights, (also called smoothing parameters), to update the components at each period.

Also known as Holt's two-parameter model

```python
from statsmodels.tsa.api import Holt

y_hat_avg = test.copy()
fit1 = Holt(np.asarray(train['Count'], dtype = np.float)).fit(smoothing_level = 0.22 , smoothing_slope = 0.16) 
#print(y_hat.head())
y_hat_avg['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(10,6))
plt.plot( train['Count'] , label = 'Train')
plt.plot(test['Count'] , label = 'Test')
plt.plot( y_hat_avg['Holt_linear'] , label = 'Holt_linear')
plt.legend(loc = 'best')
plt.title('Holt_linear')
plt.show()
```

Calculating accuracy

```python
fit1.aic
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_hat_avg.Count , y_hat_avg.Holt_linear))
print(rms)
```

# Triple exponential smoothing

Triple Exponential Smoothing is an extension of Exponential Smoothing that explicitly adds support for seasonality to the univariate time series. This method is sometimes called Holt-Winters Exponential Smoothing, named for two contributors to the method: Charles Holt and Peter Winters



```python
from statsmodels.tsa.api import ExponentialSmoothing

y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['Count'], dtype = np.float) ,seasonal_periods=7 , trend = 'add',
                           seasonal='add').fit() #we can also have 'mul'
#mul is used when we have exponential increase in dataset
#add is used when trend is increasing gradually
#We can use both the tunning technique and check aic value to compare
#seasonality is repeating after every 7 days thus we have given input as 7
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))

plt.figure(figsize=(10,6))
plt.plot( train['Count'] , label = 'Train')
plt.plot(test['Count'] , label = 'Test')
plt.plot( y_hat_avg['Holt_Winter'] , label = 'Holt_Winter')
plt.legend(loc = 'best')
plt.title('Holt_Winter')
plt.show()
```

Calculating accuracy

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_hat_avg.Count , y_hat_avg.Holt_Winter))
print(rms)
```
